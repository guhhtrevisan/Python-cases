Construa um gráfico de linhas com as seguintes características:

- Tamanho 14x6 VVVVV

- Palheta de cores Set3 VVVVV
- Destaque o impacto da pandemia (2020 e 2021) com um vspan 2020

```python
# listas de anos e refeições
anos = [2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022]
refeicoes = [40139, 127020, 168193, 153115, 202102, 232897, 277912, 205350, 233389, 232797, 110023, 160350, 222789]
```

## Tratamento de dados

```python
anos = anos
refeicoes_formatadas = ['{:,}'.format(num) for num in refeicoes]

print(anos)
print(refeicoes_formatadas)
```

## Importando bibliotecas necessárias

```python
import matplotlib.pyplot as plt
import numpy as np
from cycler import cycler
```

## Configurando paleta de cores

```python
cores = plt.get_cmap('Set3').colors
cycle_colors = cycler('color', cores)
plt.rc('axes', prop_cycle=cycle_colors)
```

## Gerando o gráfico

```python
# Marcação
marcadores = np.where((anos == np.min(anos)) | (anos == np.max(anos)), 'o', ' ')

# Gerando gráfico
plt.figure(figsize=(14,6))

for i, marcador in enumerate(marcadores):
    if marcador == 'o':
        cor = 'red'
        tamanho = 80
        plt.annotate(refeicoes_formatadas[i], color='black', xy=(anos[i], refeicoes[i]), xytext=(6,5), textcoords='offset points')
    else:
        cor = 'red'
        tamanho = 80
    plt.scatter(anos[i], refeicoes[i], marker=marcador, s=tamanho, color=cor, zorder=2)
    
plt.plot(anos, refeicoes, linewidth=3, marker='o', markersize=10, zorder=1)

plt.axvspan(2019, 2021, alpha=0.3, color=cores[3], zorder=0)
plt.text(2019.5, 75000, 'Pandemia', color='black', fontsize=11, fontweight='bold')

plt.title('Refeições servidas por ano', fontsize=16, fontweight='bold', loc='left')
plt.yticks([])
plt.xticks(anos)
plt.tick_params(axis='x', length=0)

plt.box(False)

plt.savefig('imagens/refeições_ao-ano.png', bbox_inches='tight', dpi=150)

plt.show()
```
